```python
import pandas as pd

# Load the data
df = pd.read_csv('retail_sales_dataset.csv')

# Just to check if the data is loaded
#df.head()

# ==== Discover the data =====

df.info()

# Is gender is within Male Female only?
df["Gender"].unique()

# Check if there is bias in the data
df["Gender"].value_counts()
# The result = 51% Female, 49% Male. So, there is no bias
```

```python
# Is there nulls in the data? NO
df.isna().mean()
```

```python
# Each Transaction ID is unique? YES
df["Transaction ID"].value_counts().sort_values(ascending=False).head()
```

```python
# Each Customer ID is unique? YES
df["Customer ID"].value_counts().sort_values(ascending=False).head()
```

```python
# Check if Total Amount = Quantity * Price per Unit, for every row
(df["Quantity"] * df["Price per Unit"] != df["Total Amount"]).sum()
```

```python
# Check if there is an outlier in the Age
print(df["Age"].min()) # 18
print(df["Age"].max()) # 64
# There is no outlier in the Age
```

```python
#Check if there is an outlier in Quantity
print(df["Quantity"].min())
print(df["Quantity"].max())
```

```python
# Check if there is an outliers in Price
print(df["Price per Unit"].min())
print(df["Price per Unit"].max())
```

```python
# Check if there is an outlier in Total Amount
print(df["Total Amount"].min())
print(df["Total Amount"].max())
```

```python
# Change the type of time column from str to DataTime
df["Date"]= pd.to_datetime(df["Date"])
print(df["Date"].dtype)
```

```python
# Create a new column for Month
df["Month"]= df["Date"].dt.month
print(df["Month"].unique())
```

```python
# === Sales by Month ===
# Find the month with the highest total sales revenue
df.groupby("Month")["Total Amount"].sum().sort_values(ascending=False)
```

```python
# === Top Sales product category ===
# Find the product category with the highest total sales revenue
df.groupby("Product Category")["Total Amount"].sum().sort_values(ascending=False)
```

```python
# Gender Analysis
# Find the difference in spending between males and females
df.groupby("Gender")["Total Amount"].sum().sort_values(ascending=False)
```

```python
# Identify which Age Group generates the highest sales

# Define the ranges and labels for each range
bins = [18, 25, 35, 50, 65]
labels = ["18-25", "26-35", "36-50", "50+"]

# Create new column
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True)

df["Age Group"].value_counts()
df.groupby("Age Group")["Total Amount"].sum().sort_values(ascending=False)
```

```python
# Analyze sales by age group
df.groupby(["Age Group", "Product Category"])["Total Amount"].sum().sort_values(ascending=False)
```
